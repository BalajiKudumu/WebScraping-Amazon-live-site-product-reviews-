```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url="https://www.amazon.in/Test-Exclusive-543/product-reviews/B078BN2H3R/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
page=requests.get(url)
page
```

```python
page.content
```

```python
soup=bs(page.content,"html.parser")
```

```python
print(soup.prettify())
```

```python
profiles=soup.find_all('span',class_="a-profile-name")
profiles
```

```python
cust_names=[]
for i in range(len(profiles)):
    cust_names.append(profiles[i].get_text())
    
cust_names
```

```python
cust_names.pop(0)
```

```python
cust_names.pop(1)
```

```python
cust_names
```

```python
title=soup.find_all("a", class_="review-title-content")
title
```

```python
reviews=[]
for i in range(len(title)):
    reviews.append(title[i].get_text())
reviews
```

```python
reviews[:]=[titles.lstrip("\n") for titles in reviews]
```

```python
reviews
```

```python
reviews[:]=[titles.rstrip("\n") for titles in reviews]
```

```python
reviews
```

```python
rating=soup.find_all('i', class_="review-rating")
rating
```

```python
rating_pts=[]
for pts in range(len(rating)):
    rating_pts.append(rating[pts].get_text())
rating_pts
```

```python
rating_pts.pop(0)
```

```python
rating_pts.pop(0)
```

```python
rating_pts
```

```python
msg=soup.find_all("span",{"data-hook":"review-body"})
msg
```

```python
msg_writing=[]
for i in range(len(msg)):
    msg_writing.append(msg[i].get_text())
msg_writing
```

```python
msg_writing[:]=[i.lstrip("\n") for i in msg_writing]
```

```python
msg_writing[:]=[i.rstrip("\n") for i in msg_writing]
```

```python
msg_writing
```

## Storing all those data we scraped

```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df['Customer_name']=cust_names
df["review_title"]=reviews
df['rating']=rating_pts
df['review']=msg_writing
```
